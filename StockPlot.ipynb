# StockPlot
A basic notebook fetching stock data and plotting it.

```python
import pandas as pd
import datetime 
import pandas_datareader as web
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
%matplotlib inline
```

```python
start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2015, 1, 1)
```

```python
df = web.DataReader("XOM", "iex", start, end)
```

```python
df.head()
```

```python
df['close'].plot()
```

```python
df['volume'].plot()
```

## Comparison of various plotting styles

```python
for style_name in style.available:
    style.use(style_name)
    print(style_name)
    df['close'].plot()
    plt.show()
```
